```python
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\Sayan Das\Downloads\cancer.csv")
df.head(12)
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Total.Rate', data=df)
plt.title('Boxplot of Total Cancer Rates by State')
plt.xlabel('Total Cancer Rate')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='Rates.Age.< 18', bins=30, kde=True, color='blue', label='< 18')
sns.histplot(data=df, x='Rates.Age.18-45', bins=30, kde=True, color='red', label='18-45')
sns.histplot(data=df, x='Rates.Age.45-64', bins=30, kde=True, color='green', label='45-64')
sns.histplot(data=df, x='Rates.Age.> 64', bins=30, kde=True, color='purple', label='> 64')
plt.title('Histogram of Age-wise Cancer Rates')
plt.xlabel('Cancer Rate')
plt.legend()
plt.show()
```

```python
sex_cols = [col for col in df.columns if 'Sex' in col]
sex_rates = df[sex_cols].mean()
plt.figure(figsize=(10, 6))
sns.barplot(x=sex_rates.index, y=sex_rates.values)
plt.title('Bar Plot of Sex-wise Cancer Rates')
plt.ylabel('Mean Cancer Rate')
plt.xlabel('Sex')
plt.xticks(rotation=90)
plt.show()
```

```python
race_cols = [col for col in df.columns if 'Race' in col]
race_rates = df[race_cols].mean()
plt.figure(figsize=(10, 6))
sns.barplot(x=race_rates.index, y=race_rates.values)
plt.title('Bar Plot of Race-wise Cancer Rates')
plt.ylabel('Mean Cancer Rate')
plt.xlabel('Race')
plt.xticks(rotation=90)
plt.show()
```

```python
age_cols = [col for col in df.columns if 'Rates.Age' in col]
age_rates = df[age_cols].mean()

# Create a colorful pie chart
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
plt.figure(figsize=(8, 8))
plt.pie(age_rates, labels=age_rates.index, autopct='%1.1f%%', startangle=140, colors=colors)

# Add a title
plt.title('Distribution of Cancer Rates Across Different Age Groups')

# Display the pie chart
plt.show()
```
